We need to resample the data and get rid of the missing values. Imagine we have the following dataset ( `a=[7, 9, 8, 11, 10 , 6], b=[7, 8, 10, nan, nan, nan]`) , where `a` was sampled every second and `b` was sampled every two seconds. Although `a` and `b`  look different, comparing `a` and `b` using the correct time scale shows that they are identical.

<img src="images/a_b.png" width=600/>
<center><b>a</b> and <b>b</b> match when compared on t=1,3,5</center>


Here, the missing values are added to `b` to make it the same length as `a`.

A simple approach to align `a` and `b` is to drop values at every even position in `a` and remove missing values from `b`.  


Dropping values at even positions in `a `will yield `a = [1,3,5]`. Removing missing values will result in `b` becoming `[1,3,5]`. The values are now aligned and can be compared.


Implement the strategy above to align all your data. Use any approach from Pandas to allows you to accomplish this, even if it's something we haven't covered in class.

```python
# Write your code here
a = activities.loc[:, ~activities.isna().any()]
coh1 = a.iloc[::2].reset_index(drop=True, inplace=True)

coh2 = activities.dropna()

new_activities = pd.concat([coh1, coh2], axis=1)
new_activities.shape
```

```python
# plot the Silhoutte index as function of k here.
num_clusters = [i for i in range(3,11)]
silhouette_index_scores = []

for n in num_clusters:
    cluster = KMeans(n_clusters=n, random_state=42).fit_predict(new_activities)
    silhouette_index_scores.append(silhouette_score(new_activities, cluster))
  
plt.figure(figsize=(12,6))
plt.title('Silhouette index to find optimal number of clusters')
plt.xlabel('Numer of clusters')
plt.ylabel('Silhouette index')
plt.plot(num_clusters, silhouette_index_scores)
```

#### As we can see num_cluster=3 gives the best result.

```python
# Write your code here
km_cluster = KMeans(n_clusters=3)
ypred = km_cluster.fit_predict(new_activities.T)  ## transpose because # cols is each person here.
```

* Briefly describe how you could have used Gaussian Mixture models to cluster your data.


### Answer:

The disadvantage of KMeans clustering is that it assumes the cluster shape to be circular, as it is a distance based model. It is a limiting case of Expected Maximization algorithm, which assumes an exact value of the parameter we are working with and makes it hard to work with overlapping distribution. This is where a distribution based model becomes helpful.
A Gaussian mixture model is a mixture of multiple Gaussian distributions that can be used to describe probability density functions with multiple modes, or peaks. It assumes that there are certain number of distributions, each representing a cluster. Since they are probabilistic, GMM uses a soft clustering approach.

We can solve the above problem based on GMM as follows:

```python
# Answer the quesiton above here
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=3)
gmm.fit(new_activities.T)
labels = gmm.predict(new_activities.T)

probabilities = gmm.predict_proba(new_activities.T)
```
